```python
from six import BytesIO
from PIL import Image
import numpy as np
import glob
import os
import matplotlib
import matplotlib.pyplot as plt

```

```python
def load_image(path):
  img_data=open(path,'rb').read()
  image=Image.open(BytesIO(img_data))
  (in_width,in_height)=image.size
  return np.array(image.getdata().reshape((in_height, in_width,3)).astype(np.uint8))
```

```python
elephant_image_path="/content/elephants in india"
elephant_image_np=[]
for iname in glob.glob(elephant_image_path):
  elephant_image_np.append(load_image(iname))
```

```python
plt.rcParams['figure.figsize']=[8,5]
for idx,elephant_image_np in enumerate(elephant_image_np):
  plt.subplot(2,3,idx+1)
  plt.inshow(elephant_image_np)
plt.show()
```
