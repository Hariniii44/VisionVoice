```python
import os
import tensorflow as tf
import cv2
import numpy as np
import pandas as pd
import re
from matplotlib import pyplot as plt
from scipy.sparse import csr_matrix
from scipy.sparse import vstack

# Constants
train_images_dir = '../input/flickr30k_images/flickr30k_images/flickr30k_images/'
train_captions_file = '../input/flickr30k_images/flickr30k_images/results.csv'
size = (256, 256)
num_channels = 3
bridge_size = 1024
vocab_size = 1000  # Define your vocab size

# Define the convolutional layer function
def conv_layer(inp, kernel_shape, num_kernels, num_channels, suffix):
    conv_name = f"conv{suffix}"
    with tf.name_scope(conv_name):
        kernel_init = tf.keras.initializers.RandomNormal(mean=0.0, stddev=0.01)
        conv = tf.keras.layers.Conv2D(num_kernels, kernel_shape, padding='same', activation='relu', kernel_initializer=kernel_init, name=conv_name)(inp)
    return conv

# Load image file names
train_images_list = os.listdir(train_images_dir)[:30]  # Sample size of 30
real_images = []
train = []

# Read and preprocess images
for image_name in train_images_list:
    image_path = os.path.join(train_images_dir, image_name)
    try:
        image = cv2.imread(image_path)
        resized_image = cv2.resize(image, size)
        train.append(resized_image.reshape(1, size[0], size[1], num_channels))
        real_images.append(image)
    except Exception as e:
        print(f"Error loading image {image_name}: {e}")

# Stack images vertically
train = np.vstack(train)

# Display first image
plt.imshow(np.squeeze(train[0]))
plt.show()

# Load captions
train_captions = pd.read_csv(train_captions_file, delimiter='|')
train_captions.columns = ['image_name', 'comment_number', 'comment']

def preprocess_captions(captions):
    processed_captions = []
    for caption in captions:
        if isinstance(caption, str):  # Check if the caption is a string
            caption = re.sub(' +', ' ', caption)
            caption = f'<s> {caption} <e>'
            processed_captions.append(caption.split())
        else:
            print(f"Warning: Skipping non-string caption: {caption}")
    return processed_captions



captions = preprocess_captions(train_captions['comment'].values)

# Create vocabulary
def create_vocab(captions):
    words = [word for caption in captions for word in caption]
    vocab = sorted(set(words))
    vocab_size = len(vocab)
    word_to_idx = {word: idx for idx, word in enumerate(vocab)}
    idx_to_word = {idx: word for word, idx in word_to_idx.items()}
    return vocab_size, word_to_idx, idx_to_word

vocab_size, word_to_idx, idx_to_word = create_vocab(captions)

# Convert captions to matrix representation
def captions_to_matrix(captions, vocab_size, word_to_idx):
    caption_matrices = []
    for caption in captions:
        cap_array = None
        for word in caption:
            row = [0]
            col = [word_to_idx[word]]
            data = [1]
            if cap_array is None:
                cap_array = csr_matrix((data, (row, col)), shape=(1, vocab_size))
            else:
                cap_array = vstack((cap_array, csr_matrix((data, (row, col)), shape=(1, vocab_size))))
        caption_matrices.append(cap_array)
    return caption_matrices

train_caption_matrices = captions_to_matrix(captions, vocab_size, word_to_idx)

# Define the input tensor for input images
x_inp = tf.keras.Input(shape=(size[0], size[1], num_channels), dtype=tf.float32, name='x_input')

# Then you can use x_inp in your conv_layer calls
layer_conv1 = conv_layer(inp=x_inp, kernel_shape=(3, 3), num_kernels=32, num_channels=3, suffix='1')
layer_conv2 = conv_layer(inp=layer_conv1, kernel_shape=(3, 3), num_kernels=32, num_channels=32, suffix='2')
maxpool1 = tf.nn.max_pool(layer_conv2, ksize=[1, 2, 2, 1], strides=[1, 2, 2, 1], padding='SAME')
layer_conv3 = conv_layer(inp=maxpool1, kernel_shape=(3, 3), num_kernels=64, num_channels=32, suffix='3')
layer_conv4 = conv_layer(inp=layer_conv3, kernel_shape=(3, 3), num_kernels=64, num_channels=64, suffix='4')
maxpool2 = tf.nn.max_pool(layer_conv4, ksize=[1, 2, 2, 1], strides=[1, 2, 2, 1], padding='SAME')
layer_conv5 = conv_layer(inp=maxpool2, kernel_shape=(3, 3), num_kernels=128, num_channels=64, suffix='5')
layer_conv6 = conv_layer(inp=layer_conv5, kernel_shape=(3, 3), num_kernels=128, num_channels=128, suffix='6')
maxpool3 = tf.nn.max_pool(layer_conv6, ksize=[1, 2, 2, 1], strides=[1, 2, 2, 1], padding='SAME')
layer_conv7 = conv_layer(inp=maxpool3, kernel_shape=(3, 3), num_kernels=256, num_channels=128, suffix='7')
layer_conv8 = conv_layer(inp=layer_conv7, kernel_shape=(3, 3), num_kernels=256, num_channels=256, suffix='8')

```
